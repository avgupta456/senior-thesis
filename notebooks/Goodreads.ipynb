```python
import gzip
import json
import re
import os
import sys

import numpy as np
import pandas as pd
```

```python
DIR = "../data/Goodreads/YA"
```

```python
def load_data(file_name, head = 500):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the n-th line
            if (head is not None) and (count > head):
                break
    return data
```

```python
books_path = os.path.join(DIR, "goodreads_books_young_adult.json.gz")
books_head = load_data(books_path)

reviews_path = os.path.join(DIR, "goodreads_reviews_young_adult.json.gz")
reviews_head = load_data(reviews_path)

interactions_path = os.path.join(DIR, "goodreads_interactions_young_adult.json.gz")
interactions_head = load_data(interactions_path)
```

```python
print(' == sample record (books) ==')
display(np.random.choice(books_head))
print(' == sample record (reviews) ==')
display(np.random.choice(reviews_head))
print(' == sample record (interactions) ==')
display(np.random.choice(interactions_head))
```

```python
count = 0
out = {}

with gzip.open(books_path, 'rt') as f:
    for line in f:
        count += 1
        if count % 1000 == 0:
            print(count)

        item = json.loads(line)
        out[item["book_id"]] = {
            # IDs
            "book_id": item["book_id"],
            "work_id": item["work_id"],
            "isbn": item["isbn"],
            "isbn13": item["isbn13"],
            "kindle_asin": item["kindle_asin"],
            "asin": item["asin"],
            
            # Human readable
            "title": item["title"],
            "url": item["url"],
            
            # Features
            "pub_year": -1 if item["publication_year"] == "" else int(item["publication_year"]),
            "pub_month": -1 if item["publication_month"] == "" else int(item["publication_month"]),
            "pub_day": -1 if item["publication_day"] == "" else int(item["publication_day"]),
            
            "ratings_count": -1 if item["ratings_count"] == "" else int(item["ratings_count"]),
            "text_reviews_count": -1 if item["text_reviews_count"] == "" else int(item["text_reviews_count"]),
            "average_rating": float(item["average_rating"]),
            
            "description": item["description"],
            "num_pages": -1 if item["num_pages"] == "" else int(item["num_pages"]),
            "is_ebook": item["is_ebook"],
            
            # Graph Features
            "author": item["authors"][0]["author_id"],
            "series": item["series"],
            "similar_books": item["similar_books"],
            "popular_shelves": [x for x in item["popular_shelves"] if int(x["count"]) >= 10],
            
        }
```

```python
with open("../data/Goodreads/YA/books.json", "w") as f:
    json.dump(out, f)
```
